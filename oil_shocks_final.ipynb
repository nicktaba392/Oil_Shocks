```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.io as pio
import datetime as dt
import statsmodels.api as sm
```

```python
df = pd.read_csv('df_completed2.csv')
```

```python
Oil_States = ['TX','NM','CO','AK','ND','OK']
df['oil_state'] = np.where(df['STATE_US'].isin(Oil_States),1,0)
df['big_bank'] = np.where(df['Total_Assets'] > df['Total_Assets'].median(),1,0)

df['demandxoil'] = df['Demand_Shock_right'] * df['oil_state']
df['supplyxoil'] = df['Supply_Shock_right'] * df['oil_state']
df['constant'] = 1

df['Average of Observation Value '] = df['Average of Observation Value'] /100
```

```python
res = sm.OLS( df['Nonperforming_Loans_to_Gross_Loans'],df[['Demand_Shock_right','Supply_Shock_right','VIX_right','Asset_Growth','LIQUID ASSETS / TOT. DEP. & BOR.','Equity_to_Assets','Noninterest_Expense_to_Avg_Assets','Return_on_Avg_Assets','size','Average of Observation Value']]).fit()
print(res.summary())
```
